```python
from sklearn.cross_validation import train_test_split
times = pd.DatetimeIndex(pf["date"])
group = df.groupby([times.dayofyear]).mean() 
```

```python
X = group.ix[:,3:20].values
```

```python
#Vamos a predecir el consumo energetico de los electrodomesticos asi que la variable dependeinte debe ser la columna 1
Y = group.ix[:,1].values
```

```python
#Ahora proseguimos dividiendo nuestros datos
import random
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train,Y_train)
```

```python
Y_predict = regressor.predict(X_test)
```

```python
DataY = pd.DataFrame(data = Y_predict)
```

```python
Datay = pd.DataFrame(Y_test)
```

```python
plt.scatter(x = Datay.index, y = Datay.values)
plt.scatter(x = DataY.index, y = DataY.values, c='r')
plt.title("Luces vs Electrodomesticos")
plt.xlabel("Consumo de Luces en Wh")
plt.ylabel("Consumo de electrodomesticos en Wh")
plt.show()
```
